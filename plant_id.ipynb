```python

import os
import shutil

# Define source and destination paths
src_dirs = [
    r"D:\SEM8Major\Datasets\Indian Medicinal Leaves Image Datasets\Medicinal Leaf dataset",
    r"D:\SEM8Major\Datasets\Indian Medicinal Leaves Image Datasets\Medicinal plant dataset"
]
dest_dir = r"D:\SEM8Major\Datasets\mergeddata"

# Ensure destination directory exists
os.makedirs(dest_dir, exist_ok=True)

# Function to copy contents while preserving structure
def merge_folders(src, dest):
    for root, dirs, files in os.walk(src):
        # Compute the destination path
        relative_path = os.path.relpath(root, src)
        dest_path = os.path.join(dest, relative_path)

        # Create subdirectories if they don't exist
        os.makedirs(dest_path, exist_ok=True)

        # Copy files
        for file in files:
            src_file = os.path.join(root, file)
            dest_file = os.path.join(dest_path, file)

            # Rename if file already exists
            if os.path.exists(dest_file):
                base, ext = os.path.splitext(file)
                counter = 1
                new_dest_file = os.path.join(dest_path, f"{base}_{counter}{ext}")
                while os.path.exists(new_dest_file):
                    counter += 1
                    new_dest_file = os.path.join(dest_path, f"{base}_{counter}{ext}")
                dest_file = new_dest_file

            shutil.copy2(src_file, dest_file)

# Merge both source directories into the destination
for src in src_dirs:
    merge_folders(src, dest_dir)

print("Merge completed successfully!")
```

```python
import os
import shutil

# Define the path to the data directory
data_dir = r"D:\SEM8Major\Datasets\mergeddata"

# Function to delete subfolders with less than 145 items
def delete_small_subfolders(directory, threshold=145):
    deleted_folders = []
    
    for subfolder in os.listdir(directory):
        subfolder_path = os.path.join(directory, subfolder)
        
        if os.path.isdir(subfolder_path):  # Ensure it's a folder
            num_items = len(os.listdir(subfolder_path))  # Count items
            
            if num_items < threshold:  # Delete if less than 145 items
                shutil.rmtree(subfolder_path)  # Delete folder and contents
                deleted_folders.append((subfolder, num_items))

    return deleted_folders

# Perform deletion
deleted = delete_small_subfolders(data_dir)

# Print results
if deleted:
    print("Deleted subfolders with fewer than 145 items:")
    for subfolder, count in deleted:
        print(f"Deleted: {subfolder}, Items: {count}")
else:
    print("No subfolders were deleted. All have 145 or more items.")
```

```python
import os

def rename_files_in_directory(directory):
    # Walk through each directory and its subdirectories
    for root, dirs, files in os.walk(directory):
        # Sort files to ensure sequential renaming
        files.sort()
        
        # Temporary renaming to avoid conflicts
        for index, filename in enumerate(files, start=1):
            # Check if the item is a file
            if os.path.isfile(os.path.join(root, filename)):
                # Get the file extension
                file_extension = os.path.splitext(filename)[1]
                # Define a temporary file name
                temp_name = f"temp_{index}{file_extension}"
                # Construct full file paths
                old_file = os.path.join(root, filename)
                temp_file = os.path.join(root, temp_name)
                # Rename the file to the temporary name
                os.rename(old_file, temp_file)
        
        # Final renaming from temporary to desired names
        for index, filename in enumerate(sorted(os.listdir(root)), start=1):
            # Get the file extension
            file_extension = os.path.splitext(filename)[1]
            # Define the new file name
            new_name = f"{index}{file_extension}"
            # Construct full file paths
            temp_file = os.path.join(root, filename)
            new_file = os.path.join(root, new_name)
            # Check if it's a temporary file and rename to the final name
            if filename.startswith("temp_"):
                os.rename(temp_file, new_file)
                print(f"Renamed: {temp_file} to {new_file}")

# Specify the directory
directory_path =  r"D:\SEM8Major\Datasets\mergeddata"

# Call the function to rename files
rename_files_in_directory(directory_path)
```

```python
import os
import shutil
import random

# Define directories
medi_dir = r"D:\SEM8Major\Datasets\mergeddata"
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'
val_dir = r'D:\SEM8Major\Datasets\Val'

# Split ratios
train_ratio = 0.75
val_ratio = 0.05
test_ratio = 0.20

# Ensure the destination directories exist
for dir_path in [train_dir, test_dir, val_dir]:
    os.makedirs(dir_path, exist_ok=True)

# Iterate through each subdirectory in mergeddata
for subdir in os.listdir(medi_dir):
    subdir_path = os.path.join(medi_dir, subdir)
    
    if os.path.isdir(subdir_path):  # Check if it's a directory
        images = [f for f in os.listdir(subdir_path) if os.path.isfile(os.path.join(subdir_path, f))]
        random.shuffle(images)  # Shuffle the images

        # Calculate split indices
        total_images = len(images)
        train_count = int(total_images * train_ratio)
        val_count = int(total_images * val_ratio)
        test_count = total_images - train_count - val_count

        train_images = images[:train_count]
        val_images = images[train_count:train_count + val_count]
        test_images = images[train_count + val_count:]

        # Function to move images
        def move_images(image_list, source_folder, dest_folder):
            dest_subdir = os.path.join(dest_folder, subdir)
            os.makedirs(dest_subdir, exist_ok=True)
            for image in image_list:
                shutil.move(os.path.join(source_folder, image), os.path.join(dest_subdir, image))

        # Move images to respective directories
        move_images(train_images, subdir_path, train_dir)
        move_images(val_images, subdir_path, val_dir)
        move_images(test_images, subdir_path, test_dir)

print("Data splitting completed successfully!")
```

```python
from sklearn.metrics import classification_report
import numpy as np

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
import os

# Set cache location
os.environ['TORCH_HOME'] = r'D:\SEM8Major\Cache\torch'
os.makedirs(os.environ['TORCH_HOME'], exist_ok=True)

# Device configuration
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Hyperparameters
BATCH_SIZE = 32
EPOCHS = 2

# Data transformations
train_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(30),
    transforms.RandomResizedCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices (stratified)
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.2,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / class_counts
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True
)

# Model setup
model = models.resnet18(weights='DEFAULT')
num_ftrs = model.fc.in_features

# Modified classifier
model.fc = nn.Sequential(
    nn.Linear(num_ftrs, 1024),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(1024, 512),
    nn.ReLU(),
    nn.Dropout(0.3),
    nn.Linear(512, len(class_counts))  # Use actual class count
)

model = model.to(device)

# Differential learning rates
base_params = []
new_params = []
for name, param in model.named_parameters():
    if 'fc' in name:
        new_params.append(param)
    else:
        base_params.append(param)

optimizer = optim.Adam([
    {'params': base_params, 'lr': 1e-4},
    {'params': new_params, 'lr': 1e-3}
])

# Loss with class weights
class_weights = class_weights.to(device)
criterion = nn.CrossEntropyLoss(weight=class_weights)

# Learning rate scheduler
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, 
    mode='max', 
    patience=3, 
    factor=0.5, 
    verbose=True
)

# Early stopping
class EarlyStopping:
    def __init__(self, patience=5, delta=0):
        self.patience = patience
        self.delta = delta
        self.counter = 0
        self.best_score = None
        self.early_stop = False

    def __call__(self, val_acc):
        score = val_acc
        if self.best_score is None:
            self.best_score = score
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.counter = 0

early_stopper = EarlyStopping(patience=5)

# Training loop
best_val_acc = 0.0

for epoch in range(EPOCHS):
    print(f'Epoch {epoch+1}/{EPOCHS}')
    print('-' * 10)
    
    # Training phase
    model.train()
    running_loss = 0.0
    running_corrects = 0
    
    for inputs, labels in train_loader:
        inputs = inputs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        outputs = model(inputs)
        _, preds = torch.max(outputs, 1)
        loss = criterion(outputs, labels)
        
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        
        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)
    
    epoch_loss = running_loss / len(train_subset)
    epoch_acc = running_corrects.double() / len(train_subset)
    
    print(f'Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
    
    # Validation phase
    model.eval()
    val_running_loss = 0.0
    val_running_corrects = 0
    
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs = inputs.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)
            
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            
            loss = criterion(outputs, labels)
            
            val_running_loss += loss.item() * inputs.size(0)
            val_running_corrects += torch.sum(preds == labels.data)
    
    val_epoch_loss = val_running_loss / len(val_subset)
    val_epoch_acc = val_running_corrects.double() / len(val_subset)
    
    print(f'Val Loss: {val_epoch_loss:.4f} Acc: {val_epoch_acc:.4f}\n')
    
    # Update scheduler and early stopping
    scheduler.step(val_epoch_acc)
    early_stopper(val_epoch_acc)
    
    # Save best model
    if val_epoch_acc > best_val_acc:
        best_val_acc = val_epoch_acc
        torch.save(model.state_dict(), 'best_model.pth')
        print('*** New best model saved ***\n')
    
    if early_stopper.early_stop:
        print("Early stopping triggered")
        break

# Load best model for testing
model.load_state_dict(torch.load('best_model.pth'))

# Final evaluation on test set
model.eval()
all_labels = []
all_preds = []
correct = 0
total = 0

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)
        
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        
        # Store for classification report
        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(predicted.cpu().numpy())
        
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

# Generate classification report
class_names = test_dataset.classes
report = classification_report(
    all_labels, 
    all_preds, 
    target_names=class_names,
    digits=4
)

# Save report to file
with open('classification_report.txt', 'w') as f:
    f.write(report)

# Print results to console
print("\nClassification Report:")
print(report)
print(f"Test Accuracy: {100 * correct / total:.2f}%")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import os
from tqdm.notebook import tqdm  # Notebook-optimized progress bars
import math
import optuna
from optuna.visualization import plot_optimization_history, plot_param_importances
import logging

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('improved_training.log'),
        logging.StreamHandler()
    ]
)

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters (will be optimized by Optuna)
BATCH_SIZE = 32
EPOCHS = 100
INIT_LR = 3e-4
WEIGHT_DECAY = 1e-4
DROPOUT_RATE = 0.5
GAMMA = 2.0
SMOOTHING = 0.1

# Enhanced Data Augmentation
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(45),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.GaussianBlur(3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2)
])

test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices (stratified)
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.1,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing with smoothing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True,
    num_workers=4
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

# Model Architecture
class PlantNet(nn.Module):
    def __init__(self, num_classes, dropout_rate=0.5):
        super(PlantNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            
            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            
            nn.Conv2d(512, 1024, 3, padding=1),
            nn.BatchNorm2d(1024),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d((1, 1))
        )

        self.classifier = nn.Sequential(
            nn.Dropout(dropout_rate),
            nn.Linear(1024, 2048),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(2048),
            nn.Dropout(dropout_rate),
            nn.Linear(2048, 1024),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(1024),
            nn.Dropout(dropout_rate),
            nn.Linear(1024, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

# Loss Functions
class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0, smoothing=0.1):
        super().__init__()
        self.focal = FocalLoss(alpha=alpha, gamma=gamma)
        self.label_smooth = LabelSmoothingCE(smoothing=smoothing)
        
    def forward(self, inputs, targets):
        return 0.5 * self.focal(inputs, targets) + 0.5 * self.label_smooth(inputs, targets)

class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none', weight=self.alpha)
        pt = torch.exp(-ce_loss)
        return ((1 - pt) ** self.gamma * ce_loss).mean()

class LabelSmoothingCE(nn.Module):
    def __init__(self, smoothing=0.1):
        super().__init__()
        self.smoothing = smoothing
        
    def forward(self, logits, targets):
        n_classes = logits.size(-1)
        log_preds = F.log_softmax(logits, dim=-1)
        loss = -log_preds.mean()
        nll = F.nll_loss(log_preds, targets)
        return (1 - self.smoothing) * nll + self.smoothing * loss

# Optuna Objective Function with Progress Bars
def objective(trial):
    params = {
        'batch_size': trial.suggest_categorical('batch_size', [16, 32, 64]),
        'init_lr': trial.suggest_float('init_lr', 1e-5, 1e-3, log=True),
        'weight_decay': trial.suggest_float('weight_decay', 1e-5, 1e-3, log=True),
        'dropout_rate': trial.suggest_float('dropout_rate', 0.1, 0.5),
        'gamma': trial.suggest_float('gamma', 1.0, 3.0),
        'smoothing': trial.suggest_float('smoothing', 0.05, 0.2)
    }

    model = PlantNet(len(class_counts), params['dropout_rate']).to(device)
    optimizer = optim.AdamW(model.parameters(), lr=params['init_lr'], weight_decay=params['weight_decay'])
    criterion = CombinedLoss(class_weights.to(device), params['gamma'], params['smoothing'])

    best_val_acc = 0.0
    patience_counter = 0
    
    # Main trial progress bar
    with tqdm(total=EPOCHS, desc=f'Trial {trial.number}', bar_format='{l_bar}{bar:20}{r_bar}') as trial_pbar:
        for epoch in range(EPOCHS):
            model.train()
            total_loss = 0.0
            
            # Epoch progress bar
            with tqdm(train_loader, desc=f'Epoch {epoch+1}', leave=False, 
                     bar_format='{l_bar}{bar:20}{r_bar}') as epoch_pbar:
                for inputs, labels in epoch_pbar:
                    inputs, labels = inputs.to(device), labels.to(device)
                    
                    # Mixup augmentation
                    if np.random.rand() < 0.2:
                        lam = np.random.beta(0.4, 0.4)
                        index = torch.randperm(inputs.size(0)).to(device)
                        mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
                        outputs = model(mixed_inputs)
                        loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
                    else:
                        outputs = model(inputs)
                        loss = criterion(outputs, labels)
                    
                    optimizer.zero_grad()
                    loss.backward()
                    nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                    optimizer.step()
                    
                    total_loss += loss.item()
                    epoch_pbar.set_postfix({
                        'loss': f"{total_loss/(epoch_pbar.n+1):.4f}",
                        'lr': f"{optimizer.param_groups[0]['lr']:.2e}"
                    }, refresh=False)

            # Validation
            model.eval()
            val_correct = 0
            with tqdm(val_loader, desc='Validating', leave=False, 
                     bar_format='{l_bar}{bar:20}{r_bar}') as val_pbar:
                with torch.no_grad():
                    for inputs, labels in val_pbar:
                        inputs, labels = inputs.to(device), labels.to(device)
                        outputs = model(inputs)
                        val_correct += (outputs.argmax(1) == labels).sum().item()
                        val_pbar.set_postfix({
                            'acc': f"{100*val_correct/((val_pbar.n+1)*inputs.size(0)):.2f}%"
                        }, refresh=False)

            val_acc = 100 * val_correct / len(val_subset)
            trial_pbar.set_postfix({
                'val_acc': f"{val_acc:.2f}%",
                'best': f"{best_val_acc:.2f}%"
            }, refresh=False)
            trial_pbar.update(1)

            # Early stopping
            if val_acc > best_val_acc:
                best_val_acc = val_acc
                patience_counter = 0
            else:
                patience_counter += 1
                if patience_counter >= 10:
                    break

    return best_val_acc

# Optuna Study
study = optuna.create_study(
    direction='maximize',
    storage='sqlite:///optuna_study.db',
    study_name='plantnet_optimization',
    load_if_exists=True
)
study.optimize(objective, n_trials=50, show_progress_bar=True)

# Visualization
plot_optimization_history(study).show()
plot_param_importances(study).show()

# Final Training with Best Params
best_params = study.best_params
model = PlantNet(len(class_counts), best_params['dropout_rate']).to(device)
optimizer = optim.AdamW(model.parameters(), 
                       lr=best_params['init_lr'], 
                       weight_decay=best_params['weight_decay'])
criterion = CombinedLoss(class_weights.to(device), best_params['gamma'], best_params['smoothing'])

best_val_acc = 0.0
best_model_path = 'optuna_study.pth'
epoch_pbar = tqdm(range(EPOCHS), desc='Final Training')
for epoch in epoch_pbar:
    model.train()
    train_pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}', leave=False)
    total_loss = 0.0
    
    for inputs, labels in train_pbar:
        inputs, labels = inputs.to(device), labels.to(device)
        
        # Mixup
        if np.random.rand() < 0.2:
            lam = np.random.beta(0.4, 0.4)
            index = torch.randperm(inputs.size(0)).to(device)
            mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
            outputs = model(mixed_inputs)
            loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
        
        optimizer.zero_grad()
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        
        total_loss += loss.item()
        train_pbar.set_postfix({
            'loss': f"{total_loss/(train_pbar.n+1):.4f}",
            'lr': f"{optimizer.param_groups[0]['lr']:.2e}"
        })
    
    # Validation
    model.eval()
    val_correct = 0
    val_pbar = tqdm(val_loader, desc='Validating', leave=False)
    with torch.no_grad():
        for inputs, labels in val_pbar:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            val_correct += (outputs.argmax(1) == labels).sum().item()
            val_pbar.set_postfix({
                'acc': f"{100*val_correct/((val_pbar.n+1)*inputs.size(0)):.2f}%"
            })
    
    val_acc = 100 * val_correct / len(val_subset)
    epoch_pbar.set_postfix({
        'val_acc': f"{val_acc:.2f}%",
        'best_acc': f"{best_val_acc:.2f}%"
    })
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), best_model_path)
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 10:
            print(f"\nEarly stopping at epoch {epoch+1}")
            break

# Final Evaluation
model.load_state_dict(torch.load(best_model_path))
model.eval()
test_correct = 0
test_pbar = tqdm(test_loader, desc='Testing')
with torch.no_grad():
    for inputs, labels in test_pbar:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        test_correct += (outputs.argmax(1) == labels).sum().item()
        test_pbar.set_postfix({
            'acc': f"{100*test_correct/((test_pbar.n+1)*inputs.size(0)):.2f}%"
        })

test_acc = 100 * test_correct / len(test_dataset)
print(f"\nFinal Test Accuracy: {test_acc:.2f}%")
print(classification_report(
    torch.cat([labels for _, labels in test_loader]).numpy(),
    torch.cat([model(imgs.to(device)).argmax(1).cpu() for imgs, _ in test_loader])
))
```

```python
# DO NOT EDIT
import logging
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import os
from tqdm import tqdm
import math

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('scratch_training.log'),
        logging.StreamHandler()
    ]
)

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters
BATCH_SIZE = 32
EPOCHS = 100
INIT_LR = 3e-4
WEIGHT_DECAY = 1e-4
CONFIDENCE_THRESHOLD = 0.7 

# Enhanced Data Augmentation with regularization
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(45),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.GaussianBlur(3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2)
])

# Modified test transforms with center crop
test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),  # Added to match training input size
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices (stratified)
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.1,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing with smoothing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True,
    num_workers=4
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

# Model architecture (unchanged)
class scratch(nn.Module):
    def __init__(self, num_classes):
        super(scratch, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 1024, 3, padding=1),
            nn.BatchNorm2d(1024),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d((1, 1))
        )
        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(1024, 2048),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(2048),
            nn.Dropout(0.3),
            nn.Linear(2048, 1024),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(1024),
            nn.Dropout(0.3),
            nn.Linear(1024, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

# Loss functions and training components (unchanged)
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none', weight=self.alpha)
        pt = torch.exp(-ce_loss)
        return ((1 - pt) ** self.gamma * ce_loss).mean()

class LabelSmoothingCE(nn.Module):
    def __init__(self, smoothing=0.1):
        super().__init__()
        self.smoothing = smoothing
        
    def forward(self, logits, targets):
        n_classes = logits.size(-1)
        log_preds = F.log_softmax(logits, dim=-1)
        loss = -log_preds.mean()
        nll = F.nll_loss(log_preds, targets)
        return (1 - self.smoothing) * nll + self.smoothing * loss

class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, smoothing=0.1):
        super().__init__()
        self.focal = FocalLoss(alpha=alpha)
        self.label_smooth = LabelSmoothingCE(smoothing)
        
    def forward(self, inputs, targets):
        return 0.5*self.focal(inputs, targets) + 0.5*self.label_smooth(inputs, targets)

# Initialize model and training components
model = scratch(num_classes=len(class_counts)).to(device)
optimizer = optim.AdamW(model.parameters(), lr=INIT_LR, weight_decay=WEIGHT_DECAY)
criterion = CombinedLoss(alpha=class_weights.to(device))

# Learning rate scheduler
def lr_lambda(current_step):
    warmup_steps = 500
    if current_step < warmup_steps:
        return float(current_step) / float(max(1, warmup_steps))
    return 0.5 * (1.0 + math.cos(math.pi * (current_step - warmup_steps) / (EPOCHS*len(train_loader) - warmup_steps)))

scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

# Training loop
best_val_acc = 0.0
patience_counter = 0
num_classes = len(full_train_dataset.classes)
unknown_idx = num_classes

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    train_correct = 0
    total = 0
    progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}')
    for batch_idx, (inputs, labels) in enumerate(progress_bar):
        inputs = inputs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        if np.random.rand() < 0.2:
            lam = np.random.beta(0.4, 0.4)
            index = torch.randperm(inputs.size(0)).to(device)
            mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
            outputs = model(mixed_inputs)
            loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            total += labels.size(0)
        
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        
        train_loss += loss.item() * inputs.size(0)
        progress_bar.set_postfix({'Loss': loss.item()})
    
    # Validation
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            probs = F.softmax(outputs, dim=1)
            confidences, preds = torch.max(probs, dim=1)
            preds = torch.where(confidences >= CONFIDENCE_THRESHOLD, 
                               preds, 
                               torch.tensor(unknown_idx, device=device))
            
            val_correct += (preds == labels).sum().item()
            val_loss += loss.item() * inputs.size(0)
    
    avg_train_loss = train_loss / len(train_subset)
    train_acc = 100.0 * train_correct / total if total > 0 else 0.0
    avg_val_loss = val_loss / len(val_subset)
    val_acc = 100.0 * val_correct / len(val_subset)
    
    logging.info(
        f"Epoch {epoch+1}: "
        f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'scratch.pth')
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 10:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Load best model for final evaluation and export
model.load_state_dict(torch.load('scratch.pth'))
model.eval()

# Get real test image for ONNX export
test_iter = iter(test_loader)
images, labels = next(test_iter)
sample_image = images[0].unsqueeze(0).to(device)  # Add batch dimension
sample_label = labels[0].item()

# Verify inference on sample image
with torch.no_grad():
    output = model(sample_image)
    probabilities = F.softmax(output, dim=1)
    confidence, predicted = torch.max(probabilities, 1)
    
logging.info(f"\nSample Test Image Verification:")
logging.info(f"- True label: {test_dataset.classes[sample_label]}")
logging.info(f"- Predicted: {test_dataset.classes[predicted.item()]}")
logging.info(f"- Confidence: {confidence.item():.4f}")

# Export to ONNX using real image input
torch.onnx.export(
    model,
    sample_image,
    'scratch.onnx',
    export_params=True,
    opset_version=12,
    do_constant_folding=True,
    input_names=['input'],
    output_names=['output'],
    dynamic_axes={
        'input': {0: 'batch_size'},
        'output': {0: 'batch_size'}
    }
)
logging.info("ONNX model exported successfully using real test image input!")

# Final evaluation
test_loss = 0.0
test_correct = 0
all_labels = []
all_preds = []

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        
        probs = F.softmax(outputs, dim=1)
        confidences, preds = torch.max(probs, dim=1)
        preds = torch.where(confidences >= CONFIDENCE_THRESHOLD, 
                           preds, 
                           torch.tensor(unknown_idx, device=device))
        
        test_loss += loss.item() * inputs.size(0)
        test_correct += (preds == labels).sum().item()
        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())

# Generate report
test_loss = test_loss / len(test_dataset)
test_acc = 100.0 * test_correct / len(test_dataset)
class_names = test_dataset.classes + ['unknown']

report = classification_report(
    all_labels,
    all_preds,
    target_names=class_names,
    digits=4
)

# Save results
with open('enhanced_classification_report.txt', 'w') as f:
    f.write(report)

logging.info("\nClassification Report:\n%s", report)
logging.info("Test Loss: %.4f | Test Accuracy: %.2f%%", test_loss, test_acc)

print("\nFinal Results:")
print(report)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.2f}%")
```

```python
# DO NOT EDIT
import logging
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import os
from tqdm import tqdm
import math
from PIL import Image

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('scratch_training.log'),
        logging.StreamHandler()
    ]
)

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters
BATCH_SIZE = 32
EPOCHS = 100
INIT_LR = 3e-4
WEIGHT_DECAY = 1e-4
CONFIDENCE_THRESHOLD = 0.7 

# Enhanced Data Augmentation
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(45),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.GaussianBlur(3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2)
])

test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.1,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True,
    num_workers=4
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

# Model architecture with predictor capabilities
class ScratchPredictor(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, 2),
            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),
            nn.Conv2d(512, 1024, 3, padding=1),
            nn.BatchNorm2d(1024),
            nn.ReLU(inplace=True),
            nn.AdaptiveAvgPool2d((1, 1))
        )
        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(1024, 2048),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(2048),
            nn.Dropout(0.3),
            nn.Linear(2048, 1024),
            nn.ReLU(inplace=True),
            nn.BatchNorm1d(1024),
            nn.Dropout(0.3),
            nn.Linear(1024, num_classes)
        )
        
        # Prediction components
        self.preprocess = transforms.Compose([
            transforms.Resize(256),
            transforms.CenterCrop(224),
            transforms.ToTensor(),
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])
        self.class_names = full_train_dataset.classes

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

    def predict(self, image_tensor):
        """For batch prediction"""
        with torch.no_grad():
            outputs = self.forward(image_tensor)
            probs = F.softmax(outputs, dim=1)
            confidences, class_indices = torch.max(probs, dim=1)
            return confidences, class_indices

    def predict_image(self, image_path):
        """For single image prediction"""
        image = Image.open(image_path).convert('RGB')
        tensor = self.preprocess(image).unsqueeze(0).to(device)
        confidence, class_idx = self.predict(tensor)
        
        result = {
            'confidence': confidence.item(),
            'class_index': class_idx.item(),
            'class_name': self.class_names[class_idx.item()]
        }
        
        if result['confidence'] < CONFIDENCE_THRESHOLD:
            result['class_name'] = 'unknown'
            result['class_index'] = -1
            
        return result

# Loss function
class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, smoothing=0.1):
        super().__init__()
        self.focal = FocalLoss(alpha=alpha)
        self.label_smooth = LabelSmoothingCE(smoothing)
        
    def forward(self, inputs, targets):
        return 0.5*self.focal(inputs, targets) + 0.5*self.label_smooth(inputs, targets)

# Training components
model = ScratchPredictor(num_classes=len(class_counts)).to(device)
optimizer = optim.AdamW(model.parameters(), lr=INIT_LR, weight_decay=WEIGHT_DECAY)
criterion = CombinedLoss(alpha=class_weights.to(device))

# Learning rate scheduler
def lr_lambda(current_step):
    warmup_steps = 500
    if current_step < warmup_steps:
        return float(current_step) / float(max(1, warmup_steps))
    return 0.5 * (1.0 + math.cos(math.pi * (current_step - warmup_steps) / (EPOCHS*len(train_loader) - warmup_steps)))

scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

# Training loop
best_val_acc = 0.0
patience_counter = 0

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    train_correct = 0
    total = 0
    progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}')
    
    for batch_idx, (inputs, labels) in enumerate(progress_bar):
        inputs = inputs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        # Mixed sample augmentation
        if np.random.rand() < 0.2:
            lam = np.random.beta(0.4, 0.4)
            index = torch.randperm(inputs.size(0)).to(device)
            mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
            outputs = model(mixed_inputs)
            loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            total += labels.size(0)
        
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        
        train_loss += loss.item() * inputs.size(0)
        progress_bar.set_postfix({'Loss': loss.item()})
    
    # Validation
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_loss += loss.item() * inputs.size(0)
    
    avg_train_loss = train_loss / len(train_subset)
    train_acc = 100.0 * train_correct / total if total > 0 else 0.0
    avg_val_loss = val_loss / len(val_subset)
    val_acc = 100.0 * val_correct / len(val_subset)
    
    logging.info(
        f"Epoch {epoch+1}: "
        f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model, 'scratch_predictor.pth')
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 5:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Final evaluation with proper metrics
def evaluate_predictor(model, loader):
    model.eval()
    all_labels = []
    all_preds = []
    
    with torch.no_grad():
        for inputs, labels in loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            preds = outputs.argmax(dim=1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
    
    return classification_report(
        all_labels,
        all_preds,
        target_names=model.class_names,
        digits=4,
        zero_division=0  # Add this parameter
    )

# Load best model
model = torch.load('scratch_predictor.pth', weights_only=False)
model = model.to(device)  # Move the model to the appropriate device# Generate reports
val_report = evaluate_predictor(model, val_loader)
test_report = evaluate_predictor(model, test_loader)

logging.info("\nValidation Report:\n%s", val_report)
logging.info("\nTest Report:\n%s", test_report)

# Export to ONNX
dummy_input = torch.randn(1, 3, 224, 224).to(device)
torch.onnx.export(
    model,
    dummy_input,
    'scratch_predictor.onnx',
    export_params=True,
    opset_version=12,
    do_constant_folding=True,
    input_names=['input'],
    output_names=['output'],
    dynamic_axes={
        'input': {0: 'batch_size'},
        'output': {0: 'batch_size'}
    }
)

# Example usage
sample_image_path = r"D:\SEM8Major\Datasets\testimage.jpg"
prediction = model.predict_image(sample_image_path)
print(f"\nPrediction Result for {sample_image_path}:")
print(f"Class: {prediction['class_name']}")
print(f"Confidence: {prediction['confidence']:.2%}")
print(f"Class Index: {prediction['class_index']}")

# Save class names
with open('class_names.txt', 'w') as f:
    f.write('\n'.join(model.class_names))
```

```python
# DO NOT EDIT
import logging
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import os
from tqdm import tqdm
import math
from PIL import Image

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('advanced_training.log'),
        logging.StreamHandler()
    ]
)

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters
BATCH_SIZE = 32
EPOCHS = 100
INIT_LR = 3e-4
WEIGHT_DECAY = 1e-4
CONFIDENCE_THRESHOLD = 0.7
STOCHASTIC_DEPTH_PROB = 0.2

# Enhanced Data Augmentation
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(45),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.GaussianBlur(3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2)
])

test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.1,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True,
    num_workers=4
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

# Advanced Model Components
class StochasticDepth(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p
        
    def forward(self, x):
        if not self.training:
            return x
            
        batch_size = x.size(0)
        random_tensor = torch.rand(batch_size, 1, 1, 1, device=x.device)
        keep_prob = 1 - self.p
        binary_tensor = torch.floor(random_tensor + keep_prob)
        return x * binary_tensor / keep_prob

class ResidualInceptionModule(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1):
        super().__init__()
        self.branch1 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 3, padding=1),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch2 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 5, padding=2),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch3 = nn.Sequential(
            nn.MaxPool2d(3, stride=stride, padding=1),
            nn.Conv2d(in_channels, out_channels//2, 1),
            nn.BatchNorm2d(out_channels//2),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 1, stride=stride),
                nn.BatchNorm2d(out_channels),
                StochasticDepth(STOCHASTIC_DEPTH_PROB)
            )

    def forward(self, x):
        identity = self.shortcut(x)
        branch1 = self.branch1(x)
        branch2 = self.branch2(x)
        branch3 = self.branch3(x)
        out = torch.cat([branch1, branch2, branch3], 1)
        return F.relu(out + identity)

class AdvancedScratch(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.num_classes = num_classes
        
        # Initial Stem Block
        self.stem = nn.Sequential(
            nn.Conv2d(3, 64, 3, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2, padding=1)
        )

        # Residual Inception Blocks
        self.block1 = self._make_residual_inception_block(128, 256, stride=2)
        self.block2 = self._make_residual_inception_block(256, 512, stride=2)
        self.block3 = self._make_residual_inception_block(512, 1024, stride=2)
        
        # Attention Module
        self.attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(1024, 1024//16, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024//16, 1024, 1),
            nn.Sigmoid()
        )
        
        # Final Classifier
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(1024, 2048),
            nn.BatchNorm1d(2048),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(2048, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(1024, num_classes)
        )
        
        # Initialize weights
        self._initialize_weights()

    def _make_residual_inception_block(self, in_channels, out_channels, stride):
        return nn.Sequential(
            ResidualInceptionModule(in_channels, out_channels, stride),
            ResidualInceptionModule(out_channels, out_channels, 1),
            ResidualInceptionModule(out_channels, out_channels, 1)
        )

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.stem(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        
        # Channel attention
        attention = self.attention(x)
        x = x * attention
        
        return self.classifier(x)

    def predict(self, image_tensor):
        """Batch prediction with confidence threshold"""
        with torch.no_grad():
            outputs = self.forward(image_tensor)
            probs = F.softmax(outputs, dim=1)
            confidences, class_indices = torch.max(probs, dim=1)
            return confidences, class_indices

# Loss function
class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, smoothing=0.1):
        super().__init__()
        self.focal = FocalLoss(alpha=alpha)
        self.label_smooth = LabelSmoothingCE(smoothing)
        
    def forward(self, inputs, targets):
        return 0.5*self.focal(inputs, targets) + 0.5*self.label_smooth(inputs, targets)

# Training components
model = AdvancedScratch(num_classes=len(class_counts)).to(device)
optimizer = optim.AdamW(
    model.parameters(),
    lr=INIT_LR,
    weight_decay=WEIGHT_DECAY,
    eps=1e-8
)

# Enhanced Learning Rate Schedule
def lr_lambda(current_step):
    warmup_steps = 1000
    decay_steps = EPOCHS * len(train_loader) - warmup_steps
    if current_step < warmup_steps:
        return float(current_step) / float(max(1, warmup_steps))
    progress = float(current_step - warmup_steps) / float(max(1, decay_steps))
    return 0.5 * (1.0 + math.cos(math.pi * progress))

scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

# Training loop
best_val_acc = 0.0
patience_counter = 0

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    train_correct = 0
    total = 0
    progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}')
    
    for batch_idx, (inputs, labels) in enumerate(progress_bar):
        inputs = inputs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        # Mixed sample augmentation
        if np.random.rand() < 0.2:
            lam = np.random.beta(0.4, 0.4)
            index = torch.randperm(inputs.size(0)).to(device)
            mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
            outputs = model(mixed_inputs)
            loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            total += labels.size(0)
        
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        
        train_loss += loss.item() * inputs.size(0)
        progress_bar.set_postfix({'Loss': loss.item()})
    
    # Validation
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_loss += loss.item() * inputs.size(0)
    
    avg_train_loss = train_loss / len(train_subset)
    train_acc = 100.0 * train_correct / total if total > 0 else 0.0
    avg_val_loss = val_loss / len(val_subset)
    val_acc = 100.0 * val_correct / len(val_subset)
    
    logging.info(
        f"Epoch {epoch+1}: "
        f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model, 'advanced_scratch.pth')
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 5:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Evaluation
def generate_report(model, loader):
    model.eval()
    all_labels = []
    all_preds = []
    
    with torch.no_grad():
        for inputs, labels in loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            preds = outputs.argmax(dim=1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
    
    return classification_report(
        all_labels,
        all_preds,
        target_names=full_train_dataset.classes,
        digits=4,
        zero_division=0  # Add this parameter

    )

# Load best model
model = torch.load('advanced_scratch.pth', weights_only=False)
model = model.to(device)
# Generate reports
val_report = generate_report(model, val_loader)
test_report = generate_report(model, test_loader)

logging.info("\nValidation Report:\n%s", val_report)
logging.info("\nTest Report:\n%s", test_report)

# Export to ONNX
dummy_input = torch.randn(1, 3, 224, 224).to(device)
torch.onnx.export(
    model,
    dummy_input,
    'advanced_scratch.onnx',
    export_params=True,
    opset_version=13,
    do_constant_folding=True,
    input_names=['input'],
    output_names=['output'],
    dynamic_axes={
        'input': {0: 'batch_size'},
        'output': {0: 'batch_size'}
    }
)

# Prediction Example
def predict_image(image_path):
    image = Image.open(image_path).convert('RGB')
    transform = test_transforms
    tensor = transform(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        output = model(tensor)
        probs = F.softmax(output, dim=1)
        confidence, class_idx = torch.max(probs, dim=1)
    
    class_name = full_train_dataset.classes[class_idx.item()]
    if confidence.item() < CONFIDENCE_THRESHOLD:
        class_name = "unknown"
    
    return {
        'class': class_name,
        'confidence': confidence.item(),
        'class_index': class_idx.item() if confidence.item() >= CONFIDENCE_THRESHOLD else -1
    }

# Example usage
sample_pred = predict_image(r"D:\SEM8Major\Datasets\testimage.jpg")
print(f"\nPrediction: {sample_pred}")

# Save class names
with open('class_names.txt', 'w') as f:
    f.write('\n'.join(full_train_dataset.classes))
```

```python
import logging
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
import numpy as np
import os
from tqdm import tqdm
import math
from PIL import Image

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('advanced_training.log'),
        logging.StreamHandler()
    ]
)

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters
BATCH_SIZE = 32
EPOCHS = 100
INIT_LR = 3e-4
WEIGHT_DECAY = 1e-4
STOCHASTIC_DEPTH_PROB = 0.2
TTA_NUM = 5

# Enhanced Data Augmentation
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.6, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(60),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.2),
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),
    transforms.GaussianBlur(5),
    transforms.RandomPerspective(distortion_scale=0.4, p=0.3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.5, scale=(0.02, 0.4), value='random')
])

test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.1,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True,
    num_workers=4
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

# Advanced Model Components
class StochasticDepth(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p
        
    def forward(self, x):
        if not self.training:
            return x
            
        batch_size = x.size(0)
        random_tensor = torch.rand(batch_size, 1, 1, 1, device=x.device)
        keep_prob = 1 - self.p
        binary_tensor = torch.floor(random_tensor + keep_prob)
        return x * binary_tensor / keep_prob

class ResidualInceptionModule(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1):
        super().__init__()
        self.branch1 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 3, padding=1),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch2 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 5, padding=2),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch3 = nn.Sequential(
            nn.MaxPool2d(3, stride=stride, padding=1),
            nn.Conv2d(in_channels, out_channels//2, 1),
            nn.BatchNorm2d(out_channels//2),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 1, stride=stride),
                nn.BatchNorm2d(out_channels),
                StochasticDepth(STOCHASTIC_DEPTH_PROB)
            )

    def forward(self, x):
        identity = self.shortcut(x)
        branch1 = self.branch1(x)
        branch2 = self.branch2(x)
        branch3 = self.branch3(x)
        out = torch.cat([branch1, branch2, branch3], 1)
        return F.relu(out + identity)

# Loss Functions
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none', weight=self.alpha)
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss
        return focal_loss.mean()

class LabelSmoothingCE(nn.Module):
    def __init__(self, smoothing=0.2):
        super().__init__()
        self.smoothing = smoothing

    def forward(self, inputs, targets):
        log_prob = F.log_softmax(inputs, dim=-1)
        nll_loss = -log_prob.gather(dim=-1, index=targets.unsqueeze(1))
        nll_loss = nll_loss.squeeze(1)
        smooth_loss = -log_prob.mean(dim=-1)
        loss = (1.0 - self.smoothing) * nll_loss + self.smoothing * smooth_loss
        return loss.mean()

class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, smoothing=0.2):
        super().__init__()
        self.focal = FocalLoss(alpha=alpha, gamma=2.0)
        self.label_smooth = LabelSmoothingCE(smoothing)
        
    def forward(self, inputs, targets):
        return 0.5*self.focal(inputs, targets) + 0.5*self.label_smooth(inputs, targets)

class AdvancedScratch(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.num_classes = num_classes
        
        # Initial Stem Block
        self.stem = nn.Sequential(
            nn.Conv2d(3, 64, 3, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2, padding=1)
        )

        # Residual Inception Blocks
        self.block1 = self._make_residual_inception_block(128, 256, stride=2)
        self.block2 = self._make_residual_inception_block(256, 512, stride=2)
        self.block3 = self._make_residual_inception_block(512, 1024, stride=2)
        self.block4 = self._make_residual_inception_block(1024, 2048, stride=2)
        # Attention Module
        self.attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(1024, 1024//16, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024//16, 1024, 1),
            nn.Sigmoid()
        )
        
        # Final Classifier with increased dropout
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(1024, 2048),
            nn.BatchNorm1d(2048),
            nn.ReLU(inplace=True),
            nn.Dropout(0.4),
            nn.Linear(2048, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(1024, num_classes)
        )
        
        self._initialize_weights()

    def _make_residual_inception_block(self, in_channels, out_channels, stride):
        return nn.Sequential(
            ResidualInceptionModule(in_channels, out_channels, stride),
            ResidualInceptionModule(out_channels, out_channels, 1),
        )

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.stem(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        attention = self.attention(x)
        x = x * attention
        return self.classifier(x)

class TemperatureScaling(nn.Module):
    def __init__(self, temp=1.5):
        super().__init__()
        self.temperature = nn.Parameter(torch.ones(1) * temp)

    def forward(self, logits):
        return logits / self.temperature

# Training setup
model = AdvancedScratch(num_classes=len(class_counts)).to(device)
criterion = CombinedLoss(alpha=class_weights.to(device))

optimizer = optim.AdamW(
    model.parameters(),
    lr=INIT_LR,
    weight_decay=WEIGHT_DECAY,
    eps=1e-8
)

def lr_lambda(current_step):
    warmup_steps = 1000
    decay_steps = EPOCHS * len(train_loader) - warmup_steps
    if current_step < warmup_steps:
        return float(current_step) / float(max(1, warmup_steps))
    progress = float(current_step - warmup_steps) / float(max(1, decay_steps))
    return 0.5 * (1.0 + math.cos(math.pi * progress))

scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

# Training loop
best_val_acc = 0.0
patience_counter = 0

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    train_correct = 0
    total = 0
    progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}')
    
    for batch_idx, (inputs, labels) in enumerate(progress_bar):
        inputs = inputs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        # Mixed sample augmentation
        if np.random.rand() < 0.2:
            lam = np.random.beta(0.4, 0.4)
            index = torch.randperm(inputs.size(0)).to(device)
            mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
            outputs = model(mixed_inputs)
            loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            total += labels.size(0)
        
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        
        train_loss += loss.item() * inputs.size(0)
        progress_bar.set_postfix({'Loss': loss.item()})
    
    # Validation
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_loss += loss.item() * inputs.size(0)
    
    avg_train_loss = train_loss / len(train_subset)
    train_acc = 100.0 * train_correct / total if total > 0 else 0.0
    avg_val_loss = val_loss / len(val_subset)
    val_acc = 100.0 * val_correct / len(val_subset)
    
    logging.info(
        f"Epoch {epoch+1}: "
        f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'advanced_scratch2.pth')
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 5:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Load best model
model.load_state_dict(torch.load('advanced_scratch2.pth'))
model = model.to(device)

# Temperature Scaling Calibration
temperature = TemperatureScaling().to(device)
logits_list = []
labels_list = []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        logits_list.append(model(inputs))
        labels_list.append(labels.to(device))

logits = torch.cat(logits_list)
labels = torch.cat(labels_list)

optimizer = optim.LBFGS([temperature.temperature], lr=0.01)

def closure():
    optimizer.zero_grad()
    loss = F.cross_entropy(temperature(logits), labels)
    loss.backward()
    return loss

optimizer.step(closure)

# Dynamic Confidence Threshold Calculation
confidences = []
is_correct = []
with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)
        outputs = temperature(model(inputs))  # Apply temperature
        probs = F.softmax(outputs, dim=1)
        confidence, preds = torch.max(probs, dim=1)
        confidences.extend(confidence.cpu().numpy())
        is_correct.extend((preds == labels).cpu().numpy())

thresholds = np.linspace(0.4, 0.95, 50)
f1_scores = []
for thresh in thresholds:
    filtered_preds = [c >= thresh for c in confidences]
    if sum(filtered_preds) == 0:
        f1_scores.append(0)
        continue
    f1 = f1_score(is_correct, filtered_preds)
    f1_scores.append(f1)

optimal_threshold = thresholds[np.argmax(f1_scores)]
CONFIDENCE_THRESHOLD = optimal_threshold.item()

# Enhanced Prediction Functions
def predict_image(image_path):
    image = Image.open(image_path).convert('RGB')
    tensor = test_transforms(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        output = temperature(model(tensor))  # Apply temperature
        probs = F.softmax(output, dim=1)
        confidence, class_idx = torch.max(probs, dim=1)
    
    class_name = full_train_dataset.classes[class_idx.item()]
    if confidence.item() < CONFIDENCE_THRESHOLD:
        class_name = "unknown"
    
    return {
        'class': class_name,
        'confidence': confidence.item(),
        'class_index': class_idx.item() if confidence.item() >= CONFIDENCE_THRESHOLD else -1,
        'threshold': CONFIDENCE_THRESHOLD
    }

def predict_image_with_tta(image_path):
    image = Image.open(image_path).convert('RGB')
    probs = []
    
    # Base transformation
    base_tensor = test_transforms(image).unsqueeze(0).to(device)
    with torch.no_grad():
        output = temperature(model(base_tensor))
        probs.append(F.softmax(output, dim=1))
    
    # Augmented predictions
    for _ in range(TTA_NUM-1):
        augmented_tensor = train_transforms(image).unsqueeze(0).to(device)
        with torch.no_grad():
            output = temperature(model(augmented_tensor))
            probs.append(F.softmax(output, dim=1))
    
    avg_probs = torch.mean(torch.stack(probs), dim=0)
    confidence, class_idx = torch.max(avg_probs, dim=1)
    
    class_name = full_train_dataset.classes[class_idx.item()]
    if confidence.item() < CONFIDENCE_THRESHOLD:
        class_name = "unknown"
    
    return {
        'class': class_name,
        'confidence': confidence.item(),
        'class_index': class_idx.item() if confidence.item() >= CONFIDENCE_THRESHOLD else -1,
        'threshold': CONFIDENCE_THRESHOLD,
        'tta_samples': TTA_NUM
    }

# Generate reports
def generate_report(model, loader, temperature):
    model.eval()
    all_labels = []
    all_preds = []
    
    with torch.no_grad():
        for inputs, labels in loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = temperature(model(inputs))
            preds = outputs.argmax(dim=1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
    
    return classification_report(
        all_labels,
        all_preds,
        target_names=full_train_dataset.classes,
        digits=4,
        zero_division=0
    )

# Final evaluation
val_report = generate_report(model, val_loader, temperature)
test_report = generate_report(model, test_loader, temperature)

logging.info("\nValidation Report (Calibrated):\n%s", val_report)
logging.info("\nTest Report (Calibrated):\n%s", test_report)

# Error Analysis
class ErrorAnalyzer:
    def __init__(self, class_names):
        self.errors = []
        self.class_names = class_names
    
    def add(self, image_tensor, true_class, pred_class, confidence):
        self.errors.append({
            'image': image_tensor.cpu(),
            'true': self.class_names[true_class],
            'pred': self.class_names[pred_class],
            'confidence': confidence
        })

# Example usage
analyzer = ErrorAnalyzer(full_train_dataset.classes)
sample_pred = predict_image_with_tta(r"D:\SEM8Major\Datasets\testimage.jpg")
print(f"\nCalibrated Prediction: {sample_pred}")

# Save class names
with open('class_names.txt', 'w') as f:
    f.write('\n'.join(full_train_dataset.classes))
```

```python
import logging
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
import numpy as np
import os
from tqdm import tqdm
import math
from PIL import Image

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('advanced_training.log'),
        logging.StreamHandler()
    ]
)

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters
BATCH_SIZE = 16
EPOCHS = 100
INIT_LR = 3e-4
WEIGHT_DECAY = 1e-4
STOCHASTIC_DEPTH_PROB = 0.2
TTA_NUM = 5

# Enhanced Data Augmentation
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.6, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(60),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.2),
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),
    transforms.GaussianBlur(5),
    transforms.RandomPerspective(distortion_scale=0.4, p=0.3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.5, scale=(0.02, 0.4), value='random')
])

test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.1,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True,
    num_workers=4
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

# Advanced Model Components
class StochasticDepth(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p
        
    def forward(self, x):
        if not self.training:
            return x
            
        batch_size = x.size(0)
        random_tensor = torch.rand(batch_size, 1, 1, 1, device=x.device)
        keep_prob = 1 - self.p
        binary_tensor = torch.floor(random_tensor + keep_prob)
        return x * binary_tensor / keep_prob

class ResidualInceptionModule(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1):
        super().__init__()
        self.branch1 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 3, padding=1),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch2 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 5, padding=2),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch3 = nn.Sequential(
            nn.MaxPool2d(3, stride=stride, padding=1),
            nn.Conv2d(in_channels, out_channels//2, 1),
            nn.BatchNorm2d(out_channels//2),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 1, stride=stride),
                nn.BatchNorm2d(out_channels),
                StochasticDepth(STOCHASTIC_DEPTH_PROB)
            )

    def forward(self, x):
        identity = self.shortcut(x)
        branch1 = self.branch1(x)
        branch2 = self.branch2(x)
        branch3 = self.branch3(x)
        out = torch.cat([branch1, branch2, branch3], 1)
        return F.relu(out + identity)

# Loss Functions
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none', weight=self.alpha)
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss
        return focal_loss.mean()

class LabelSmoothingCE(nn.Module):
    def __init__(self, smoothing=0.2):
        super().__init__()
        self.smoothing = smoothing

    def forward(self, inputs, targets):
        log_prob = F.log_softmax(inputs, dim=-1)
        nll_loss = -log_prob.gather(dim=-1, index=targets.unsqueeze(1))
        nll_loss = nll_loss.squeeze(1)
        smooth_loss = -log_prob.mean(dim=-1)
        loss = (1.0 - self.smoothing) * nll_loss + self.smoothing * smooth_loss
        return loss.mean()

class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, smoothing=0.2):
        super().__init__()
        self.focal = FocalLoss(alpha=alpha, gamma=2.0)
        self.label_smooth = LabelSmoothingCE(smoothing)
        
    def forward(self, inputs, targets):
        return 0.5*self.focal(inputs, targets) + 0.5*self.label_smooth(inputs, targets)

class AdvancedScratch(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.num_classes = num_classes
        
        # Initial Stem Block
        self.stem = nn.Sequential(
            nn.Conv2d(3, 64, 3, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2, padding=1)
        )

        # Residual Inception Blocks
        self.block1 = self._make_residual_inception_block(128, 256, stride=2)
        self.block2 = self._make_residual_inception_block(256, 512, stride=2)
        self.block3 = self._make_residual_inception_block(512, 1024, stride=2)
        # Attention Module
        self.attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(1024, 1024//16, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024//16, 1024, 1),
            nn.Sigmoid()
        )
        
        # Final Classifier with increased dropout
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(1024, 2048),
            nn.BatchNorm1d(2048),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(2048, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(1024, num_classes)
        )
        
        self._initialize_weights()

    def _make_residual_inception_block(self, in_channels, out_channels, stride):
        return nn.Sequential(
            ResidualInceptionModule(in_channels, out_channels, stride),
            ResidualInceptionModule(out_channels, out_channels, 1),
        )

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.stem(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        attention = self.attention(x)
        x = x * attention
        return self.classifier(x)

class TemperatureScaling(nn.Module):
    def __init__(self, temp=1.5):
        super().__init__()
        self.temperature = nn.Parameter(torch.ones(1) * temp)

    def forward(self, logits):
        return logits / self.temperature

# Training setup
model = AdvancedScratch(num_classes=len(class_counts)).to(device)
criterion = CombinedLoss(alpha=class_weights.to(device))

optimizer = optim.AdamW(
    model.parameters(),
    lr=INIT_LR,
    weight_decay=WEIGHT_DECAY,
    eps=1e-8
)

def lr_lambda(current_step):
    warmup_steps = 1000
    decay_steps = EPOCHS * len(train_loader) - warmup_steps
    if current_step < warmup_steps:
        return float(current_step) / float(max(1, warmup_steps))
    progress = float(current_step - warmup_steps) / float(max(1, decay_steps))
    return 0.5 * (1.0 + math.cos(math.pi * progress))

scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

# Training loop
best_val_acc = 0.0
patience_counter = 0

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    train_correct = 0
    total = 0
    progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}')
    
    for batch_idx, (inputs, labels) in enumerate(progress_bar):
        inputs = inputs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        # Mixed sample augmentation
        if np.random.rand() < 0.2:
            lam = np.random.beta(0.4, 0.4)
            index = torch.randperm(inputs.size(0)).to(device)
            mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
            outputs = model(mixed_inputs)
            loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            total += labels.size(0)
        
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        
        train_loss += loss.item() * inputs.size(0)
        progress_bar.set_postfix({'Loss': loss.item()})
    
    # Validation
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_loss += loss.item() * inputs.size(0)
    
    avg_train_loss = train_loss / len(train_subset)
    train_acc = 100.0 * train_correct / total if total > 0 else 0.0
    avg_val_loss = val_loss / len(val_subset)
    val_acc = 100.0 * val_correct / len(val_subset)
    
    logging.info(
        f"Epoch {epoch+1}: "
        f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'advanced_scratch2.pth')
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 5:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Load best model
model.load_state_dict(torch.load('advanced_scratch2.pth'))
model = model.to(device)

# Temperature Scaling Calibration
temperature = TemperatureScaling().to(device)
logits_list = []
labels_list = []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        logits_list.append(model(inputs))
        labels_list.append(labels.to(device))

logits = torch.cat(logits_list)
labels = torch.cat(labels_list)

optimizer = optim.LBFGS([temperature.temperature], lr=0.01)

def closure():
    optimizer.zero_grad()
    loss = F.cross_entropy(temperature(logits), labels)
    loss.backward()
    return loss

optimizer.step(closure)

# Dynamic Confidence Threshold Calculation
confidences = []
is_correct = []
with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)
        outputs = temperature(model(inputs))  # Apply temperature
        probs = F.softmax(outputs, dim=1)
        confidence, preds = torch.max(probs, dim=1)
        confidences.extend(confidence.cpu().numpy())
        is_correct.extend((preds == labels).cpu().numpy())

thresholds = np.linspace(0.4, 0.95, 50)
f1_scores = []
for thresh in thresholds:
    filtered_preds = [c >= thresh for c in confidences]
    if sum(filtered_preds) == 0:
        f1_scores.append(0)
        continue
    f1 = f1_score(is_correct, filtered_preds)
    f1_scores.append(f1)

optimal_threshold = thresholds[np.argmax(f1_scores)]
CONFIDENCE_THRESHOLD = optimal_threshold.item()

# Enhanced Prediction Functions
def predict_image(image_path):
    image = Image.open(image_path).convert('RGB')
    tensor = test_transforms(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        output = temperature(model(tensor))  # Apply temperature
        probs = F.softmax(output, dim=1)
        confidence, class_idx = torch.max(probs, dim=1)
    
    class_name = full_train_dataset.classes[class_idx.item()]
    if confidence.item() < CONFIDENCE_THRESHOLD:
        class_name = "unknown"
    
    return {
        'class': class_name,
        'confidence': confidence.item(),
        'class_index': class_idx.item() if confidence.item() >= CONFIDENCE_THRESHOLD else -1,
        'threshold': CONFIDENCE_THRESHOLD
    }

def predict_image_with_tta(image_path):
    image = Image.open(image_path).convert('RGB')
    probs = []
    
    # Base transformation
    base_tensor = test_transforms(image).unsqueeze(0).to(device)
    with torch.no_grad():
        output = temperature(model(base_tensor))
        probs.append(F.softmax(output, dim=1))
    
    # Augmented predictions
    for _ in range(TTA_NUM-1):
        augmented_tensor = train_transforms(image).unsqueeze(0).to(device)
        with torch.no_grad():
            output = temperature(model(augmented_tensor))
            probs.append(F.softmax(output, dim=1))
    
    avg_probs = torch.mean(torch.stack(probs), dim=0)
    confidence, class_idx = torch.max(avg_probs, dim=1)
    
    class_name = full_train_dataset.classes[class_idx.item()]
    if confidence.item() < CONFIDENCE_THRESHOLD:
        class_name = "unknown"
    
    return {
        'class': class_name,
        'confidence': confidence.item(),
        'class_index': class_idx.item() if confidence.item() >= CONFIDENCE_THRESHOLD else -1,
        'threshold': CONFIDENCE_THRESHOLD,
        'tta_samples': TTA_NUM
    }

# Generate reports
def generate_report(model, loader, temperature):
    model.eval()
    all_labels = []
    all_preds = []
    
    with torch.no_grad():
        for inputs, labels in loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = temperature(model(inputs))
            preds = outputs.argmax(dim=1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
    
    return classification_report(
        all_labels,
        all_preds,
        target_names=full_train_dataset.classes,
        digits=4,
        zero_division=0
    )

# Final evaluation
val_report = generate_report(model, val_loader, temperature)
test_report = generate_report(model, test_loader, temperature)

logging.info("\nValidation Report (Calibrated):\n%s", val_report)
logging.info("\nTest Report (Calibrated):\n%s", test_report)

# Error Analysis
class ErrorAnalyzer:
    def __init__(self, class_names):
        self.errors = []
        self.class_names = class_names
    
    def add(self, image_tensor, true_class, pred_class, confidence):
        self.errors.append({
            'image': image_tensor.cpu(),
            'true': self.class_names[true_class],
            'pred': self.class_names[pred_class],
            'confidence': confidence
        })

# Example usage
analyzer = ErrorAnalyzer(full_train_dataset.classes)
sample_pred = predict_image_with_tta(r"D:\SEM8Major\Datasets\testimage.jpg")
print(f"\nCalibrated Prediction: {sample_pred}")

# Save class names
with open('class_names.txt', 'w') as f:
    f.write('\n'.join(full_train_dataset.classes))
```

```python
import logging
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
import numpy as np
import os
from tqdm import tqdm
import math
from PIL import Image

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('advanced_training.log'),
        logging.StreamHandler()
    ]
)

# Define paths
train_dir = r'D:\SEM8Major\Datasets\Train'
test_dir = r'D:\SEM8Major\Datasets\Test'

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Hyperparameters
BATCH_SIZE = 64
EPOCHS = 100
INIT_LR = 3e-4
WEIGHT_DECAY = 1e-4
STOCHASTIC_DEPTH_PROB = 0.2
TTA_NUM = 5

# Enhanced Data Augmentation
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.6, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(60),
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.2),
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),
    transforms.GaussianBlur(5),
    transforms.RandomPerspective(distortion_scale=0.4, p=0.3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.5, scale=(0.02, 0.4), value='random')
])

test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load datasets
full_train_dataset = datasets.ImageFolder(train_dir, train_transforms)
test_dataset = datasets.ImageFolder(test_dir, test_transforms)

# Split indices
train_indices, val_indices = train_test_split(
    list(range(len(full_train_dataset))),
    test_size=0.1,
    stratify=full_train_dataset.targets,
    random_state=42
)

# Create subsets
train_subset = Subset(full_train_dataset, train_indices)
val_subset = Subset(full_train_dataset, val_indices)

# Class balancing
class_counts = torch.bincount(torch.tensor(full_train_dataset.targets))
class_weights = 1. / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(full_train_dataset.targets)[train_indices]]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data loaders
train_loader = DataLoader(
    train_subset,
    batch_size=BATCH_SIZE,
    sampler=sampler,
    pin_memory=True,
    num_workers=4
)

val_loader = DataLoader(
    val_subset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

test_loader = DataLoader(
    test_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    pin_memory=True,
    num_workers=4
)

# Advanced Model Components
class StochasticDepth(nn.Module):
    def __init__(self, p=0.5):
        super().__init__()
        self.p = p
        
    def forward(self, x):
        if not self.training:
            return x
            
        batch_size = x.size(0)
        random_tensor = torch.rand(batch_size, 1, 1, 1, device=x.device)
        keep_prob = 1 - self.p
        binary_tensor = torch.floor(random_tensor + keep_prob)
        return x * binary_tensor / keep_prob

class ResidualInceptionModule(nn.Module):
    def __init__(self, in_channels, out_channels, stride=1):
        super().__init__()
        self.branch1 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 3, padding=1),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch2 = nn.Sequential(
            nn.Conv2d(in_channels, out_channels//4, 1, stride=stride),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels//4, out_channels//4, 5, padding=2),
            nn.BatchNorm2d(out_channels//4),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.branch3 = nn.Sequential(
            nn.MaxPool2d(3, stride=stride, padding=1),
            nn.Conv2d(in_channels, out_channels//2, 1),
            nn.BatchNorm2d(out_channels//2),
            nn.ReLU(inplace=True),
            StochasticDepth(STOCHASTIC_DEPTH_PROB)
        )
        
        self.shortcut = nn.Sequential()
        if stride != 1 or in_channels != out_channels:
            self.shortcut = nn.Sequential(
                nn.Conv2d(in_channels, out_channels, 1, stride=stride),
                nn.BatchNorm2d(out_channels),
                StochasticDepth(STOCHASTIC_DEPTH_PROB)
            )

    def forward(self, x):
        identity = self.shortcut(x)
        branch1 = self.branch1(x)
        branch2 = self.branch2(x)
        branch3 = self.branch3(x)
        out = torch.cat([branch1, branch2, branch3], 1)
        return F.relu(out + identity)

# Loss Functions
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none', weight=self.alpha)
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss
        return focal_loss.mean()

class LabelSmoothingCE(nn.Module):
    def __init__(self, smoothing=0.2):
        super().__init__()
        self.smoothing = smoothing

    def forward(self, inputs, targets):
        log_prob = F.log_softmax(inputs, dim=-1)
        nll_loss = -log_prob.gather(dim=-1, index=targets.unsqueeze(1))
        nll_loss = nll_loss.squeeze(1)
        smooth_loss = -log_prob.mean(dim=-1)
        loss = (1.0 - self.smoothing) * nll_loss + self.smoothing * smooth_loss
        return loss.mean()

class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, smoothing=0.2):
        super().__init__()
        self.focal = FocalLoss(alpha=alpha, gamma=2.0)
        self.label_smooth = LabelSmoothingCE(smoothing)
        
    def forward(self, inputs, targets):
        return 0.5*self.focal(inputs, targets) + 0.5*self.label_smooth(inputs, targets)

class AdvancedScratch(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.num_classes = num_classes
        
        # Initial Stem Block
        self.stem = nn.Sequential(
            nn.Conv2d(3, 64, 3, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2, padding=1)
        )

        # Residual Inception Blocks
        self.block1 = self._make_residual_inception_block(128, 256, stride=2)
        self.block2 = self._make_residual_inception_block(256, 512, stride=2)
        self.block3 = self._make_residual_inception_block(512, 1024, stride=2)
        # Attention Module
        self.attention = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(1024, 1024//16, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(1024//16, 1024, 1),
            nn.Sigmoid()
        )
        
        # Final Classifier with increased dropout
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d((1, 1)),
            nn.Flatten(),
            nn.Linear(1024, 2048),
            nn.BatchNorm1d(2048),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(2048, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(1024, num_classes)
        )
        
        self._initialize_weights()

    def _make_residual_inception_block(self, in_channels, out_channels, stride):
        return nn.Sequential(
            ResidualInceptionModule(in_channels, out_channels, stride),
            ResidualInceptionModule(out_channels, out_channels, 1),
        )

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)

    def forward(self, x):
        x = self.stem(x)
        x = self.block1(x)
        x = self.block2(x)
        x = self.block3(x)
        attention = self.attention(x)
        x = x * attention
        return self.classifier(x)

class TemperatureScaling(nn.Module):
    def __init__(self, temp=1.5):
        super().__init__()
        self.temperature = nn.Parameter(torch.ones(1) * temp)

    def forward(self, logits):
        return logits / self.temperature

# Training setup
model = AdvancedScratch(num_classes=len(class_counts)).to(device)
criterion = CombinedLoss(alpha=class_weights.to(device))

optimizer = optim.AdamW(
    model.parameters(),
    lr=INIT_LR,
    weight_decay=WEIGHT_DECAY,
    eps=1e-8
)

def lr_lambda(current_step):
    warmup_steps = 1000
    decay_steps = EPOCHS * len(train_loader) - warmup_steps
    if current_step < warmup_steps:
        return float(current_step) / float(max(1, warmup_steps))
    progress = float(current_step - warmup_steps) / float(max(1, decay_steps))
    return 0.5 * (1.0 + math.cos(math.pi * progress))

scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

# Training loop
best_val_acc = 0.0
patience_counter = 0

for epoch in range(EPOCHS):
    model.train()
    train_loss = 0.0
    train_correct = 0
    total = 0
    progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}')
    
    for batch_idx, (inputs, labels) in enumerate(progress_bar):
        inputs = inputs.to(device, non_blocking=True)
        labels = labels.to(device, non_blocking=True)
        
        optimizer.zero_grad()
        
        # Mixed sample augmentation
        if np.random.rand() < 0.2:
            lam = np.random.beta(0.4, 0.4)
            index = torch.randperm(inputs.size(0)).to(device)
            mixed_inputs = lam * inputs + (1 - lam) * inputs[index]
            outputs = model(mixed_inputs)
            loss = lam * criterion(outputs, labels) + (1 - lam) * criterion(outputs, labels[index])
        else:
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            train_correct += (preds == labels).sum().item()
            total += labels.size(0)
        
        loss.backward()
        nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        scheduler.step()
        
        train_loss += loss.item() * inputs.size(0)
        progress_bar.set_postfix({'Loss': loss.item()})
    
    # Validation
    model.eval()
    val_loss = 0.0
    val_correct = 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            preds = outputs.argmax(dim=1)
            val_correct += (preds == labels).sum().item()
            val_loss += loss.item() * inputs.size(0)
    
    avg_train_loss = train_loss / len(train_subset)
    train_acc = 100.0 * train_correct / total if total > 0 else 0.0
    avg_val_loss = val_loss / len(val_subset)
    val_acc = 100.0 * val_correct / len(val_subset)
    
    logging.info(
        f"Epoch {epoch+1}: "
        f"Train Loss: {avg_train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
        f"Val Loss: {avg_val_loss:.4f} | Val Acc: {val_acc:.2f}%"
    )
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), 'advanced_scratch2.pth')
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 5:
            print(f"Early stopping at epoch {epoch+1}")
            break

# Load best model
model.load_state_dict(torch.load('advanced_scratch2.pth'))
model = model.to(device)

# Temperature Scaling Calibration
temperature = TemperatureScaling().to(device)
logits_list = []
labels_list = []

with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        logits_list.append(model(inputs))
        labels_list.append(labels.to(device))

logits = torch.cat(logits_list)
labels = torch.cat(labels_list)

optimizer = optim.LBFGS([temperature.temperature], lr=0.01)

def closure():
    optimizer.zero_grad()
    loss = F.cross_entropy(temperature(logits), labels)
    loss.backward()
    return loss

optimizer.step(closure)

# Dynamic Confidence Threshold Calculation
confidences = []
is_correct = []
with torch.no_grad():
    for inputs, labels in val_loader:
        inputs = inputs.to(device)
        labels = labels.to(device)
        outputs = temperature(model(inputs))  # Apply temperature
        probs = F.softmax(outputs, dim=1)
        confidence, preds = torch.max(probs, dim=1)
        confidences.extend(confidence.cpu().numpy())
        is_correct.extend((preds == labels).cpu().numpy())

thresholds = np.linspace(0.4, 0.95, 50)
f1_scores = []
for thresh in thresholds:
    filtered_preds = [c >= thresh for c in confidences]
    if sum(filtered_preds) == 0:
        f1_scores.append(0)
        continue
    f1 = f1_score(is_correct, filtered_preds)
    f1_scores.append(f1)

optimal_threshold = thresholds[np.argmax(f1_scores)]
CONFIDENCE_THRESHOLD = optimal_threshold.item()

# Enhanced Prediction Functions
def predict_image(image_path):
    image = Image.open(image_path).convert('RGB')
    tensor = test_transforms(image).unsqueeze(0).to(device)
    
    with torch.no_grad():
        output = temperature(model(tensor))  # Apply temperature
        probs = F.softmax(output, dim=1)
        confidence, class_idx = torch.max(probs, dim=1)
    
    class_name = full_train_dataset.classes[class_idx.item()]
    if confidence.item() < CONFIDENCE_THRESHOLD:
        class_name = "unknown"
    
    return {
        'class': class_name,
        'confidence': confidence.item(),
        'class_index': class_idx.item() if confidence.item() >= CONFIDENCE_THRESHOLD else -1,
        'threshold': CONFIDENCE_THRESHOLD
    }

def predict_image_with_tta(image_path):
    image = Image.open(image_path).convert('RGB')
    probs = []
    
    # Base transformation
    base_tensor = test_transforms(image).unsqueeze(0).to(device)
    with torch.no_grad():
        output = temperature(model(base_tensor))
        probs.append(F.softmax(output, dim=1))
    
    # Augmented predictions
    for _ in range(TTA_NUM-1):
        augmented_tensor = train_transforms(image).unsqueeze(0).to(device)
        with torch.no_grad():
            output = temperature(model(augmented_tensor))
            probs.append(F.softmax(output, dim=1))
    
    avg_probs = torch.mean(torch.stack(probs), dim=0)
    confidence, class_idx = torch.max(avg_probs, dim=1)
    
    class_name = full_train_dataset.classes[class_idx.item()]
    if confidence.item() < CONFIDENCE_THRESHOLD:
        class_name = "unknown"
    
    return {
        'class': class_name,
        'confidence': confidence.item(),
        'class_index': class_idx.item() if confidence.item() >= CONFIDENCE_THRESHOLD else -1,
        'threshold': CONFIDENCE_THRESHOLD,
        'tta_samples': TTA_NUM
    }

# Generate reports
def generate_report(model, loader, temperature):
    model.eval()
    all_labels = []
    all_preds = []
    
    with torch.no_grad():
        for inputs, labels in loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = temperature(model(inputs))
            preds = outputs.argmax(dim=1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
    
    return classification_report(
        all_labels,
        all_preds,
        target_names=full_train_dataset.classes,
        digits=4,
        zero_division=0
    )

# Final evaluation
val_report = generate_report(model, val_loader, temperature)
test_report = generate_report(model, test_loader, temperature)

logging.info("\nValidation Report (Calibrated):\n%s", val_report)
logging.info("\nTest Report (Calibrated):\n%s", test_report)

# Error Analysis
class ErrorAnalyzer:
    def __init__(self, class_names):
        self.errors = []
        self.class_names = class_names
    
    def add(self, image_tensor, true_class, pred_class, confidence):
        self.errors.append({
            'image': image_tensor.cpu(),
            'true': self.class_names[true_class],
            'pred': self.class_names[pred_class],
            'confidence': confidence
        })

# Example usage
analyzer = ErrorAnalyzer(full_train_dataset.classes)
sample_pred = predict_image_with_tta(r"D:\SEM8Major\Datasets\testimage.jpg")
print(f"\nCalibrated Prediction: {sample_pred}")

# Save class names
with open('class_names.txt', 'w') as f:
    f.write('\n'.join(full_train_dataset.classes))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, WeightedRandomSampler
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import json
import os
import math  # Added missing import
from PIL import Image 
# Configuration
class Config:
    train_dir = r'D:\SEM8Major\Datasets\Train'
    val_dir = r'D:\SEM8Major\Datasets\Val'
    test_dir = r'D:\SEM8Major\Datasets\Test'
    batch_size = 32
    epochs = 100
    lr = 3e-4
    weight_decay = 1e-4
    patience = 5
    model_save_path = "plantnet_model.pth"
    class_to_idx_save = "class_to_idx.json"
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Data Transforms
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(256, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(45),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.GaussianBlur(3),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2)
])

val_test_transforms = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load Datasets
train_dataset = datasets.ImageFolder(Config.train_dir, transform=train_transforms)
val_dataset = datasets.ImageFolder(Config.val_dir, transform=val_test_transforms)
test_dataset = datasets.ImageFolder(Config.test_dir, transform=val_test_transforms)

# Save class to index mapping
with open(Config.class_to_idx_save, 'w') as f:
    json.dump(train_dataset.class_to_idx, f)

# Class Balancing
class_counts = torch.bincount(torch.tensor(train_dataset.targets))
class_weights = 1.0 / (class_counts + 1e-4)
samples_weights = class_weights[torch.tensor(train_dataset.targets)]
sampler = WeightedRandomSampler(samples_weights, len(samples_weights), replacement=True)

# Data Loaders
train_loader = DataLoader(train_dataset, batch_size=Config.batch_size, sampler=sampler, 
                         num_workers=4, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=Config.batch_size, shuffle=False, 
                       num_workers=4, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=Config.batch_size, shuffle=False,
                        num_workers=4, pin_memory=True)
```

```python

# Model Architecture
class PlantNet(nn.Module):
    def __init__(self, num_classes):
        super(PlantNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),
            
            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(),
            
            nn.Conv2d(512, 1024, 3, padding=1),
            nn.BatchNorm2d(1024),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1))
        )
        
        self.classifier = nn.Sequential(
            nn.Dropout(0.5),
            nn.Linear(1024, 2048),
            nn.ReLU(),
            nn.BatchNorm1d(2048),
            nn.Dropout(0.3),
            nn.Linear(2048, 1024),
            nn.ReLU(),
            nn.BatchNorm1d(1024),
            nn.Dropout(0.3),
            nn.Linear(1024, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

# Loss Function
class CombinedLoss(nn.Module):
    def __init__(self, alpha=None, smoothing=0.1):
        super().__init__()
        self.alpha = alpha.to(Config.device) if alpha is not None else None
        self.smoothing = smoothing

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = ((1 - pt) ** 2 * ce_loss).mean()
        
        n_classes = inputs.size(-1)
        log_preds = F.log_softmax(inputs, dim=-1)
        nll = F.nll_loss(log_preds, targets)
        smooth_loss = -log_preds.mean() * self.smoothing
        label_smooth_loss = (1 - self.smoothing) * nll + smooth_loss
        
        return 0.5 * focal_loss + 0.5 * label_smooth_loss

# Training Setup
model = PlantNet(num_classes=len(class_counts)).to(Config.device)
optimizer = optim.AdamW(model.parameters(), lr=Config.lr, weight_decay=Config.weight_decay)
criterion = CombinedLoss(alpha=class_weights, smoothing=0.1)

def lr_lambda(current_step):
    warmup_steps = 500
    if current_step < warmup_steps:
        return current_step / warmup_steps
    return 0.5 * (1 + math.cos(math.pi * (current_step - warmup_steps) / (Config.epochs*len(train_loader) - warmup_steps)))

scheduler = optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

# Training Loop
train_losses, val_losses = [], []
train_accs, val_accs = [], []
best_val_acc = 0.0
patience_counter = 0

for epoch in range(Config.epochs):
    # Training
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{Config.epochs}"):
        images, labels = images.to(Config.device), labels.to(Config.device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        scheduler.step()
        
        running_loss += loss.item()
        _, predicted = outputs.max(1)
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()
    
    train_loss = running_loss/len(train_loader)
    train_acc = 100.*correct/total
    train_losses.append(train_loss)
    train_accs.append(train_acc)

    # Validation
    model.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(Config.device), labels.to(Config.device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            val_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()
    
    val_loss = val_loss/len(val_loader)
    val_acc = 100.*correct/total
    val_losses.append(val_loss)
    val_accs.append(val_acc)
    
    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
    print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
    
    # Early Stopping & Model Saving
    if val_acc > best_val_acc:
        print(f"Validation accuracy improved from {best_val_acc:.2f}% to {val_acc:.2f}%")
        best_val_acc = val_acc
        torch.save({
            'model_state_dict': model.state_dict(),
            'class_to_idx': train_dataset.class_to_idx
        }, Config.model_save_path)
        patience_counter = 0
    else:
        patience_counter += 1
        print(f"Patience counter: {patience_counter}/{Config.patience}")
        if patience_counter >= Config.patience:
            print("Early stopping triggered!")
            break

```
